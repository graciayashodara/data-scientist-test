```python
import pandas as pd
import json
import numpy as np
```

```python
with open('data-sample (1).json') as file:
    data = json.load(file)
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv('data.csv', index=False)
```

```python
dt = df.copy()
```

```python
# Check for NaN or non-dictionary values in 'taskLocationDone' column
mask = dt['taskLocationDone'].apply(lambda x: isinstance(x, dict) if not pd.isnull(x) else False)

# Extract longitude and latitude values from 'taskLocationDone' column
dt.loc[mask, 'lon'] = dt.loc[mask, 'taskLocationDone'].apply(lambda x: x['lon'])
dt.loc[mask, 'lat'] = dt.loc[mask, 'taskLocationDone'].apply(lambda x: x['lat'])

# Fill NaN values in the new columns with a default value (e.g., 0)
dt['lon'] = dt['lon'].fillna(0)
dt['lat'] = dt['lat'].fillna(0)

# Drop the original 'taskLocationDone' column
dt = dt.drop('taskLocationDone', axis=1)
```

```python
dt.head()
```

```python
# Check for NaN or non-dictionary values in 'cod' column
mask = dt['cod'].apply(lambda x: isinstance(x, dict) if not pd.isnull(x) else False)

# Extract 'amount' and 'received' values from 'cod' column
dt.loc[mask, 'cod_amount'] = dt.loc[mask, 'cod'].apply(lambda x: x['amount'])
dt.loc[mask, 'cod_status_received'] = dt.loc[mask, 'cod'].apply(lambda x: x['received'])

# Fill NaN values in the new columns with a default value (e.g., False for cod_status and 0 for cod_amount)
dt['cod_status_received'] = dt['cod_status_received'].fillna(False)
dt['cod_amount'] = dt['cod_amount'].fillna(0)

# Drop the original 'cod' column
dt = dt.drop('cod', axis=1)
```

```python
dt.head()
```

```python
dt.loc[0, 'UserVar']
```

```python
import numpy as np

# Check for NaN or non-dictionary values in 'UserVar' column
mask = dt['UserVar'].apply(lambda x: isinstance(x, dict) if not pd.isnull(x) else False)

# Extract values from 'UserVar' column
dt.loc[mask, 'branch_dest'] = dt.loc[mask, 'UserVar'].apply(lambda x: x.get('branch_dest'))
dt.loc[mask, 'taskStatusLabel'] = dt.loc[mask, 'UserVar'].apply(lambda x: x.get('taskStatusLabel'))
dt.loc[mask, 'receiver_city'] = dt.loc[mask, 'UserVar'].apply(lambda x: x.get('receiver_city'))
dt.loc[mask, 'taskDetailStatusLabel'] = dt.loc[mask, 'UserVar'].apply(lambda x: x.get('taskDetailStatusLabel'))
dt.loc[mask, 'taskDetailStatus'] = dt.loc[mask, 'UserVar'].apply(lambda x: x.get('taskDetailStatus'))
dt.loc[mask, 'weight'] = dt.loc[mask, 'UserVar'].apply(lambda x: x.get('weight'))
dt.loc[mask, 'branch_origin'] = dt.loc[mask, 'UserVar'].apply(lambda x: x.get('branch_origin'))
dt.loc[mask, 'taskStatus'] = dt.loc[mask, 'UserVar'].apply(lambda x: x.get('taskStatus'))

# Fill NaN values in the new columns with np.nan
dt['branch_dest'] = dt['branch_dest'].fillna(np.nan)
dt['taskStatusLabel'] = dt['taskStatusLabel'].fillna(np.nan)
dt['receiver_city'] = dt['receiver_city'].fillna(np.nan)
dt['taskDetailStatusLabel'] = dt['taskDetailStatusLabel'].fillna(np.nan)
dt['taskDetailStatus'] = dt['taskDetailStatus'].fillna(np.nan)
dt['weight'] = dt['weight'].fillna(np.nan)
dt['branch_origin'] = dt['branch_origin'].fillna(np.nan)
dt['taskStatus'] = dt['taskStatus'].fillna(np.nan)

# Drop the original 'UserVar' column
dt = dt.drop('UserVar', axis=1)
```

```python
dt
```

```python
dt['taskCreatedTime'] = dt['taskCreatedTime'].str.replace(' \+0700', '')
dt['taskCreatedTime'] = pd.to_datetime(dt['taskCreatedTime'], format='%Y-%m-%d %H:%M:%S')
```

```python
dt['taskCompletedTime'] = dt['taskCompletedTime'].str.replace(' \+0700', '')
dt['taskCompletedTime'] = dt['taskCompletedTime'].str.replace(' \+08:00', '')
dt['taskCompletedTime'] = dt['taskCompletedTime'].str.replace(' \+0800', '')
dt['taskCompletedTime'] = pd.to_datetime(dt['taskCompletedTime'], format='%Y-%m-%d %H:%M:%S')
```
